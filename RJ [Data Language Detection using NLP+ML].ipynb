```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("language.csv")
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer    #Convert a collection of text documents to a matrix of token counts.

# create a CountVectorizer Object
vectorizer = CountVectorizer()

# sample test data
data = ["love data science ","love machine learning"]

#fit and transform the data
vectorized_data = vectorizer.fit_transform(data)
#Get vocabulary (unique words)
print (vectorizer.get_feature_names_out())

#convert the result to an Array
print(vectorized_data.toarray())
```

```python
from sklearn.model_selection import train_test_split #Split arrays or matrices into random train and test subsets.
```

```python
from sklearn.naive_bayes import MultinomialNB # multinomial Naive Bayes classifier is suitable for classification with discrete features 
                                               #  (e.g., word counts for text classification)
```

```python
x = np.array(data['Text'])
y = np.array(data['language'])
```

```python
cv = CountVectorizer()
x = cv.fit_transform(x) # Machine Learning Model 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state = 42)
```

```python
x_train
```

```python
print(x_train)
```

```python
y_test
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test) # model  accuracy 
```

```python
user = input("enter a Text")
data = cv.transform([user]).toarray()
output = model.predict(data)
print (output)
```
